```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
sns.set()
```

```python
rnd_num = np.random.RandomState(42)
X = np.dot(rnd_num.rand(2,2), rnd_num.randn(2, 500)).T
```

```python
X.shape
X
```

```python
X[:, 0] = - X[:, 0]
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print (pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)

#plot data

for k, v in zip(pca.explained_variance_, pca.components_):
    vec = v * 3 * np.sqrt(k)
    
    ax = plt.gca()
    arrowprops=dict(arrowstyle='<-', 
                    linewidth=1,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', pca.mean_, pca.mean_, + vec, arrowprops=arrowprops)
    ax.text(-0.90, 1.2, 'PC1', ha = 'center', va = 'center', rotation =-42, size=12)
    ax.text(-0.1, 0.6, 'PC2', ha = 'center', va = 'center', rotation =50, size=12)
plt.axis('equal')
```

# Dimensionality Reduction with PCA

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
X.shape
```

```python
X_pca.shape
```

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal')
```
